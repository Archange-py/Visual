```python
A, B, C, D = Point(0,0,"A"), Point(1,1,"B"), Point(2,2,"C"), Point(3,3,"D")

liste0 = [A, B, C, D]

liste1 = connect_points(liste0, ending=False)
liste2 = connect_points(liste0, ending=True)

for i in liste0: print(i.name, end=' ')
print()
for i in liste1: print((i[0].name, i[1].name), end=' ')
print()
for i in liste2: print((i[0].name, i[1].name), end=' ')
```

#### *interpolate*

```python
yellow = (255,255,0)
red = (255,0,0)

print(interpolate(yellow, red, 10))

def draw_interpolate(C1, C2, N, position=(0, 0), taille=1):
  colors_interpolate = interpolate(C1, C2, N)
  for n, col in enumerate(colors_interpolate):
    fill_rect(position[0]+n*taille, position[1], taille, taille, col)
  del colors_interpolate

#for y in range(222):
#  draw_interpolate(yellow, red, N=320, position=(0, y), taille=1)
```
